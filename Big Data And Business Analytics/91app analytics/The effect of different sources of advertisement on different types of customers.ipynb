```python
import os
data=os.listdir('C://Users//POPO//Downloads//91ForNTUDataSet//123_new')
purchase=[]
traffic =[]
for i in data:
    if "traffic" in i:
        traffic.append(i)
    if "purchase" in i:
        purchase.append(i)
print(purchase[:5])
print(traffic[:5])
```

```python
import pandas as pd
df1=[pd.read_csv(os.path.join("C://Users//POPO//Downloads//91ForNTUDataSet//123_new",i)) for i in purchase]
df2=[pd.read_csv(os.path.join("C://Users//POPO//Downloads//91ForNTUDataSet//123_new",i)) for i in traffic]
```

```python
df_purchase=pd.concat(df1,axis=0)
df_traffic=pd.concat(df2,axis=0)
df_traffic.head()
```

```python
df_traffic=df_traffic.dropna(subset=["uid"])
df_traffic=df_traffic.drop_duplicates()
df_purchase=df_purchase.dropna(subset=["uid"])
df_purchase=df_purchase.drop_duplicates()
print(df_traffic.shape)
print(df_purchase.shape)
```

```python
import datetime
df_traffic["ht"]=df_traffic["ht"].apply(lambda x:datetime.date.fromtimestamp(int(x) / 1000).strftime("%Y-%m-%d"))
df_purchase["ht"]=df_purchase["ht"].apply(lambda x:datetime.date.fromtimestamp(int(x) / 1000).strftime("%Y-%m-%d"))
```

```python
df_purchase.head()
```

```python
buy= df_purchase.groupby('uid').ht.apply(list).to_dict()
print(buy["YP4rsCTcLZMhO8CXsQDoB5KonLHEM6km1ycEBT9Hxtw%3D"])
print(len(buy))
```

```python
#前10 大廣告來源 佔了整體廣告活動來源的96%
print("%.2f" % (pd.value_counts(df_traffic["cs"])[:10].sum()/pd.value_counts(df_traffic["cs"]).sum()))
print(pd.value_counts(df_traffic["cs"])[:10])
```

```python
default=5
ad={}
pur={}
for i in pd.unique(df_traffic["uid"]):
    ad[i]=[0]*10
    pur[i]=[0]*10
top_10=list(pd.value_counts(df_traffic["cs"])[:10].index)
for i in range(len(df_traffic)):
    source=df_traffic.iloc[i,8]
    uid=df_traffic.iloc[i,3]
    time=df_traffic.iloc[i,4]
    if source in top_10:
        index=top_10.index(source)
        ad[uid][index]+=1
        for i in range(default):
            delta = datetime.timedelta(days = i) 
            time_index=(pd.to_datetime(time)+delta).strftime("%Y-%m-%d")
            if uid in buy.keys():
                if time_index in buy[uid]:
                    pur[uid][index]+=1
```

```python
get_ad= pd.DataFrame([v for v in ad.values()], columns=top_10,index=ad.keys())

real_buy=pd.DataFrame([v for v in pur.values()], columns=top_10,index=pur.keys())
```

```python
get_ad.head() 
```

```python
#real_buy.head()
```

```python
final=real_buy.div(get_ad)
RFM=pd.read_csv("rfm.csv")
final=final.reset_index()
final=pd.merge(final,RFM,left_on="index",right_on="MemberID",how="left")
final=final.dropna(subset=["Cluster"])
final["Cluster"]=final["Cluster"].apply(lambda x:int(x)).astype("category")
final.head()
```

```python
final["Cluster"].value_counts()
```

```python
result=final.groupby("Cluster")[top_10].mean()
result
```
