```python
checkin=["郭台銘","鴻海","夏普","收購","面板","富連網","高橋興三","富士康","台銘","簽約","日圓","成交值","組裝","大立光","鴻夏戀","社長"
       ,"蘋果","電子","戴正吳","台積電","日本","權值","董事長","集團"]
def check(x):
    for i in checkin:
        if i in str(x):
            return 1
            break
        else:
            return 0
    
bbs["target"]=bbs["concat"].apply(check)
forum["target"]=forum["concat"].apply(check)
news["target"]=news["concat"].apply(check)
```

```python
def judge2016(x):
    if x >0 and x>0.007552:
        return "rise"
    elif x<0 and x<-0.004998:
        return "fall"
    else:
        return "same"
import numpy as np
def judge2017(x):
    if x >0 and x>np.percentile(stock2017[stock2017["漲跌"]>0]["漲跌"],25):
        return "rise"
    elif x<=0 and x<=np.percentile(stock2017[stock2017["漲跌"]<=0]["漲跌"],75):
        return "fall"
    else:
        return "same"
def judge2018(x):
    if x >0 and x>np.percentile(stock2018[stock2018["漲跌"]>0]["漲跌"],25):
        return "rise"
    elif x<=0 and x<=np.percentile(stock2018[stock2018["漲跌"]<=0]["漲跌"],75):
        return "fall"
    else:
        return "same"
stock2016["看漲看跌"]=stock2016["漲跌"].apply(judge2016)
stock2017["看漲看跌"]=stock2017["漲跌"].apply(judge2017)
stock2018["看漲看跌"]=stock2018["漲跌"].apply(judge2018)
```

```python
ans2016=[]
for i in range(len(data2016)):
    if data2016["weekday"].iloc[i]==0 or data2016["weekday"].iloc[i]==1 or data2016["weekday"].iloc[i]==2:
        check=data2016["post_time"].iloc[i]+timedelta(days=3)
        ans2016.append(stock2016[stock2016["年月日"]==check]["看漲看跌"].values)
    elif data2016["weekday"].iloc[i]==3 or data2016["weekday"].iloc[i]==4 or data2016["weekday"].iloc[i]==5:
        check=data2016["post_time"].iloc[i]+timedelta(days=5)
        ans2016.append(stock2016[stock2016["年月日"]==check]["看漲看跌"].values) 
    else:
        check=data2016["post_time"].iloc[i]+timedelta(days=4)
        ans2016.append(stock2016[stock2016["年月日"]==check]["看漲看跌"].values)

data2016["tag"]=ans2016   
```
