```python
import pandas as pd

data=pd.read_csv("key word.csv",engine="python")
data.head()
```

```python
total=pd.read_csv("Total key word.csv",engine="python")
```

```python
total.head()
```

```python
#該主題的文件
doc=5896
import numpy as np
data["TF-IDF"]=(1+np.log10(data["TF"]))*(np.log10(doc/data["DF"]))
```

```python
#全部文件
total_doc=90507
total["TF-IDF"]=(1+np.log10(total["TF"]))*(np.log10(total_doc/total["DF"]))
```

```python
TF=total[["詞","TF"]].set_index("詞").to_dict()
TF=TF["TF"]
DF=total[["詞","DF"]].set_index("詞").to_dict()
DF=DF["DF"]
TFIDF=total[["詞","TF-IDF"]].set_index("詞").to_dict()
TFIDF=TFIDF["TF-IDF"]
```

```python
data["全部TF"]=data["詞"].map(TF)
data["全部DF"]=data["詞"].map(DF)
data["全部TF-IDF"]=data["詞"].map(TFIDF)

data["TF期望值"]=(data["全部TF"]*(doc/total_doc)).apply(lambda x:int(x))
data["DF期望值"]=(data["全部DF"]*(doc/total_doc)).apply(lambda x:int(x))
```

```python
data["TF卡方值(保留正負號)"]=(np.sign(data["TF"]-data["TF期望值"])*(data["TF"]-data["TF期望值"])**2/(data["TF期望值"]))
data["DF卡方值(保留正負號)"]=(np.sign(data["DF"]-data["DF期望值"])*(data["DF"]-data["DF期望值"])**2/(data["DF期望值"]))
```

```python
data["lift"]=(data["DF"]/doc)/(data["全部DF"]/total_doc)
```

```python
data["MI"]=np.log10(data["DF"]/(data["全部DF"]*doc))
data.head()
```
