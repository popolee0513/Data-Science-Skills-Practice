```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
datas = pd.read_csv('C://Users//POPO//Desktop//Carseats.csv')
datas = datas.drop(datas.keys()[0], axis = 1) 

# 計算相關性
co = datas.corr()

# 繪製熱點圖
mask = np.zeros_like(co)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(co, mask=mask, vmax=.3, square=True)
```

```python
Age = datas['Age']
tick_arr = np.arange(0,100,5) 

ax = sns.distplot(Age, bins =10, kde=False )
ax.set(title = 'Number of Age', xlabel='Age', 
       ylabel = 'Number', xlim = [0,100], 
       xticks=tick_arr )
```
