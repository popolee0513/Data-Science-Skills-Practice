```python
from sklearn.cluster import AgglomerativeClustering
import pandas as pd
import seaborn as sns
```

```python
HW=pd.read_csv("C://Users//POPO//Desktop//DATA_2.01_SKU.csv")
from sklearn.preprocessing import StandardScaler  #標準化數據
scaler = StandardScaler()
HW_copy = HW.copy()
scaler.fit(HW_copy)
HW_copy=scaler.transform(HW_copy)
```

```python
clustering = AgglomerativeClustering(linkage='ward',affinity='euclidean',n_clusters=3).fit(HW_copy)
```

```python
cluster_labels=clustering.labels_
```

```python
HW=HW.assign(Cluster = cluster_labels)
```

```python
sns.lmplot(x='CV',y='ADS',data=HW,fit_reg=False,hue="Cluster")
```
