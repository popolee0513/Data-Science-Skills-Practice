```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
def get_spin_result(win_prob):  
    result = False 
    if np.random.random() <= win_prob:  
        result = True  
    return result 
def test_code(i): 
    win_prob = 18/38 # set appropriately to the probability of a win 
    np.random.seed(i) # do this only once 
    #print(get_spin_result(win_prob)) # test the roulette spin  
        # add your code here to implement the experiments
    episode_winnings = 0
    count=0
    winnings=[0]
    while episode_winnings < 80 and count<1000 :
        won = False
        bet_amount = 1
        count+=1
        while not won:
            won = get_spin_result(win_prob)
            if won == True:
                episode_winnings = episode_winnings + bet_amount
                winnings.append(episode_winnings)
            else:
                episode_winnings = episode_winnings - bet_amount
                bet_amount = bet_amount * 2
                winnings.append(episode_winnings)
    while len(winnings)<1000 :
        final=winnings[-1]
        if 80 in winnings:
            winnings.append(80)
        else:
            winnings.append(final)
    return np.array(winnings)

if __name__ == "__main__": 
    fig, ax = plt.subplots()
    ax.set_xlim(0,300)
    ax.set_ylim(-256,100)
    for i in range(10):
        win=test_code(i)
        x=np.array(range(1000))
        ax.plot(x, win, label="trial_"+str(i))
    ax.legend(loc='best')
    plt.show()
  
```

```python
if __name__ == "__main__": 
    fig, ax = plt.subplots()
    ax.set_xlim(0,300)
    ax.set_ylim(-256,100)
    ans=[]
    for i in range(1000):
        win=test_code(i)
        ans.append(win)
    a1=np.mean(ans,axis=0)
    a2=np.std(ans,axis=0)
    a3=a1+a2
    a4=a1-a2
    
    ax.plot(np.array(range(1000)),a1,label="mean")
    ax.plot(np.array(range(1000)),a3,label="mean+std")
    ax.plot(np.array(range(1000)),a4,label="mean-std")
    ax.legend(loc='best')
    plt.show()
```
